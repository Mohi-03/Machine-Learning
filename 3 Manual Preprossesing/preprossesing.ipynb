#### Read the csv file

```python
import pandas as pd
df = pd.read_csv("Cars93.csv", na_values=["", "NA"], keep_default_na=False)
df.head()
```

#### Step-3 Replace missing values

##### 2.Continuous (Numeric) - Mean, Median


##### Fillna function to replace missing values

```python
m[m>0]
```

```python
md=df["AirBags"].mode()[0]
md
```

```python
df["AirBags"] = df["AirBags"].fillna(md)
```

```python
df.head()
```

```python
df.columns
```

```python
df.dtypes
```

```python

cat = list(df.columns[df.dtypes == "object"])
cat
```

```python
con = list(df.columns[df.dtypes != "object"])
con
```

```python
def replacer(df):
    cat = list(df.columns[df.dtypes == "object"])
    for i in df.columns:
        if i in cat:
            md = df[i].mode()[0]
            df[i] = df[i].fillna(md)
        else:
            mn = df[i].mean()
            df[i] = df[i].fillna(mn)
    
    # Print that values replaced
    print(f"Missing values replaced")
```

```python
replacer(df)
```

```python
df.isna().sum()
```

```python
df.to_csv("cars93Clean.csv",index=False)
```

#### Setp 3 - Swperate X and Y (Weight)
##### 1. Weight ~ remaing features
##### 2. id is just a serial number column - Statstically insignificant 

```python

X = df.drop(columns=["id", "Weight"])
Y = df[["Weight"]]
X.head()
```

```python
Y.head()
```

#### Setp 4 - Apply Preprocessing on X
##### 1.Categorical - OneHotEncoding
##### 2. Continuous - StandardScaler


```python
X_cat = X.select_dtypes(include="object")
X_cat.head()
```

```python
X_cat["Cylinders"].unique()
```

```python
X_con = X.select_dtypes(exclude="object")
X_con
```

##### One Hot Encoding should be applied on categorical columns

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(
    handle_unknown="ignore", sparse_output=False).set_output(transform="pandas")
X_cat_pre = ohe.fit_transform(X_cat)
X_cat_pre.head()
```

#### Continuous columns apply standard scaler
##### mean = 0, stdev = 1, Your numerical data will be in same range

```python
X_con.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().set_output(transform="pandas")
X_con_pre = scaler.fit_transform(X_con)
X_con_pre.head()
```


#### Combine cat and con preprocessed data

```python
X_pre = X_con_pre.join(X_cat_pre)
X_pre.head()
```

```python

X_pre.to_csv("Cars93Pre.csv", index=False)
```

#### Train Test Splits

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.2, random_state=21)
```

#### Model Building

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(xtrain,ytrain)
```

#### Evaluate r2 score

```python

model.score(xtrain, ytrain)
```

```python

model.score(xtest, ytest)
```

```python
from sklearn.metrics import (
    root_mean_squared_error,
    mean_absolute_error,
    mean_absolute_percentage_error,
    r2_score
)
```

```python
def evaluate_model(model, x, y):
    # Predict the results
    ypred = model.predict(x)
    # Evalutate the metrics
    rmse = root_mean_squared_error(y, ypred)
    mae = mean_absolute_error(y, ypred)
    mape = mean_absolute_percentage_error(y, ypred)
    r2 = r2_score(y, ypred)
    # Print your results
    print(f"RMSE : {rmse:.2f}")
    print(f"MAE : {mae:.2f}")
    print(f"MAPE : {mape:.4f}")
    print(f"R2 : {r2:.4f}")
```

```python
print("Training Results :")
evaluate_model(model,xtrain,ytrain)
```

```python
ypread_train = model.predict(xtrain)
ypread_train[0:5]
```

```python
# testing results
print("Testing Results :")
evaluate_model(model,xtest,ytest)
```
